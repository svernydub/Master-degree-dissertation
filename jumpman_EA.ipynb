```python
import matplotlib.pyplot as plt
import seaborn as sns  
import numpy as np
import pandas as pd
```

ENGAGEMENT ANALYSIS

```python
#Loading urls and data-likes and comments


# Posts before co-branding
before_co_branding_posts = {
    "h1": (48923, 89,  "03.12.2019", "Before Co-Branding_id_1"),
    "h2": (119730, 209, "30.11.2019", "Before Co-Branding_id_2"),
    "h3": (41638, 57, "15.11.2019", "Before Co-Branding_id_3"),
    "h4": (74203, 243, "14.11.2019", "Before Co-Branding_id_4"),
    "h5": (60948, 83, "14.11.2019", "Before Co-Branding_id_5")
}

# Main posts
co_branding_posts = {
    "h6": (113153, 308, "04.12.2019", "Co-Branding_id_1"),
    "h7": (259667, 706, "19.01.2020", "Co-Branding_id_2")
}

# Posts after co-branding
after_co_branding_posts = {
    "h8": (314637, 565, "23.01.2020", "After Co-Branding_id_1"),
    "h9": (283348, 435, "03.02.2020", "After Co-Branding_id_2"),
    "h10": (91710, 121, "14.02.2020", "After Co-Branding_id_3"),
    "h11": (195193, 219, "15.02.2020", "After Co-Branding_id_4"),
    "h12": (152570, 187, "16.02.2020", "After Co-Branding_id_5")
}

posts = { **before_co_branding_posts, **co_branding_posts, **after_co_branding_posts}

# Create DataFrame
posts = pd.DataFrame.from_dict(posts,  orient='index', columns=['Likes', "Comments", 'Date', 'Post'])


posts
```

```python
# Combine all posts
all_posts = {**co_branding_posts, **before_co_branding_posts, **after_co_branding_posts}

# Create DataFrame
df_main = pd.DataFrame.from_dict(all_posts, orient='index', columns=['Likes', "Comments", 'Date', 'Category'])
df_main.index.name = 'URL'

# Convert 'Date' column to datetime
df_main['Date'] = pd.to_datetime(df_main['Date'], format='%d.%m.%Y')

# Reset index to move 'URL' to a column
df_main.reset_index(inplace=True)

# Reorder the columns
df_main = df_main[['URL', 'Date', 'Category', 'Likes', 'Comments']]

df_main
```

```python
#Loading urls and data-likes and comments


before_co_branding_posts_plot = {
    "h1": (48923, 89,  "03.12.2019", "Before Co-Branding"),
    "h2": (119730, 209, "30.11.2019", "Before Co-Branding"),
    "h3": (41638, 57, "15.11.2019", "Before Co-Branding"),
    "h4": (74203, 243, "14.11.2019", "Before Co-Branding"),
    "h5": (60948, 83, "14.11.2019", "Before Co-Branding")
}

# Main posts
co_branding_posts_plot = {
    "h6": (113153, 308, "04.12.2019", "Co-Branding"),
    "h7": (259667, 706, "19.01.2020", "Co-Branding")
}

# Posts after co-branding
after_co_branding_posts_plot = {
    "h8": (314637, 565, "23.01.2020", "After Co-Branding"),
    "h9": (283348, 435, "03.02.2020", "After Co-Branding"),
    "h10": (91710, 121, "14.02.2020", "After Co-Branding"),
    "h11": (195193, 219, "15.02.2020", "After Co-Branding"),
    "h12": (152570, 187, "16.02.2020", "After Co-Branding")
}
posts_plot = { **before_co_branding_posts_plot, **co_branding_posts_plot, **after_co_branding_posts_plot}

# Create DataFrame
posts_plot = pd.DataFrame.from_dict(posts_plot,  orient='index', columns=['Likes', "Comments", 'Date', 'Post'])
posts_plot
```

```python
# Plotting
plt.figure(figsize=(10,6))

# Plotting likes
bars = plt.bar(posts['Post'], posts['Likes'], color='grey', label='Likes')

# Adding labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             '%d' % int(height), ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Post')
plt.ylabel('Count')
plt.title('Likes for Different Posts')
plt.xticks(rotation=45)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

# Plotting likes
bars = plt.bar(posts['Post'], posts['Comments'], color='grey', label='Comments')

# Adding labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             '%d' % int(height), ha='center', va='bottom')

# Adding labels and title
plt.xlabel('Post')
plt.ylabel('Count')
plt.title('Comments for Different Posts')
plt.xticks(rotation=45)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Calculate likes, comments per category
df_plot = posts_plot.groupby('Post')[['Likes', 'Comments']].sum()

# Define the desired order of categories
desired_order = ['Before Co-Branding', 'Co-Branding', 'After Co-Branding']

# Sort the total_engagement_per_category DataFrame based on the desired order
df_plot = df_plot.reindex(desired_order)

df_plot
```

```python
posts_plot
```

Likes Analysis

```python
# Plot the sum Likes per category
ax = df_plot['Likes'].plot(kind='bar', color='grey')
plt.title('Total Likes Per Category')
plt.xlabel('Post')
plt.ylabel('Likes')
plt.xticks(rotation=45)


# Add value labels to the bars
for i, v in enumerate(df_plot['Likes']):
    ax.text(i, v + 4000, str(v), ha='center', va='bottom')

plt.show()
```

```python
# Calculate for each category
mean_likes = posts_plot.groupby('Post')['Likes'].agg(['mean'])
# Create a DataFrame with relevant data for comparison
comparison_likes = mean_likes.loc[['Before Co-Branding', 'Co-Branding', 'After Co-Branding']]

comparison_likes
```

```python
# Reordering the DataFrame based on the desired order of categories
desired_order = ['Before Co-Branding', 'Co-Branding', 'After Co-Branding']
mean_likes = mean_likes.reindex(desired_order)

# Plotting
ax = mean_likes.plot(kind='bar',color ="grey", figsize=(10, 6))
plt.title('Mean Likes for Each Category')
plt.xlabel('Category')
plt.ylabel('Likes')
plt.xticks(rotation=45)
plt.legend(title='Statistic')
for p in ax.patches:
    ax.annotate(f"{int(p.get_height())}", 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), 
                textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python

mean_likes = {
    'Before Co-Branding': 69088,
    'Co-Branding': 186410,
    'After Co-Branding': 207491
}

# Calculate the change in mean comments from 'Before Co-Branding' to 'Co-Branding'
change_comments_before_co_branding_percent = ((mean_likes['Co-Branding'] - mean_likes['Before Co-Branding']) / mean_likes['Before Co-Branding']) * 100
change_comments_co_before_branding_absolute = mean_likes['Co-Branding'] - mean_likes['Before Co-Branding']

# Calculate the change in mean comments  from 'Co-Branding' to 'After Co-Branding'
change_comments_after_co_branding_percent = ((mean_likes['After Co-Branding'] - mean_likes['Co-Branding']) / mean_likes['Co-Branding']) * 100
change_comments_after_co_branding_absolute = mean_likes['After Co-Branding'] - mean_likes['Co-Branding']

print("Change in mean likes from 'Before Co-Branding' to 'Co-Branding': {:.2f}% ({:.2f} comments)".format(change_comments_before_co_branding_percent, change_comments_co_before_branding_absolute))
print("Change in mean likes from 'Co-Branding' to 'After Co-Branding': {:.2f}% ({:.2f} comments)".format(change_comments_after_co_branding_percent, change_comments_after_co_branding_absolute))
```

```python
# PLOTTING LIKES CHANGE

# Calculate the changes in mean likes
changes_percent_likes = {
    'Before to Co-Branding': ((mean_likes['Co-Branding'] - mean_likes['Before Co-Branding']) / mean_likes['Before Co-Branding']) * 100,
    'Co-Branding to After': ((mean_likes['After Co-Branding'] - mean_likes['Co-Branding']) / mean_likes['Co-Branding']) * 100
}

changes_absolute_likes = {
    'Before Co-Branding': mean_likes['Co-Branding'] - mean_likes['Before Co-Branding'],
    'After Co-Branding': mean_likes['After Co-Branding'] - mean_likes['Co-Branding']
}

# Plotting
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Bar plot for percentage changes
ax[0].bar(changes_percent_likes.keys(), changes_percent_likes.values(), color=['grey', 'grey'])
ax[0].set_ylabel('Percentage Change')
ax[0].set_title('Percentage Change in Mean Likes')
for i, v in enumerate(changes_percent_likes.values()):
    ax[0].text(i, v, f"{v:.2f}%", ha='center', va='bottom')

# Bar plot for absolute changes
ax[1].bar(changes_absolute_likes.keys(), changes_absolute_likes.values(), color=['grey', 'grey'])
ax[1].set_ylabel('Absolute Change')
ax[1].set_title('Absolute Change in Mean Likes')
for i, v in enumerate(changes_absolute_likes.values()):
    ax[1].text(i, v, f"{v:.2f}", ha='center', va='bottom')

# Show plot
plt.tight_layout()
plt.show()
```
